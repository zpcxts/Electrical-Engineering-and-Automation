# 低压断路器保护线路长度计算方法

当线路上的电流（长时间或大量）超过导线的载流能力时，会烧毁电路。为了避免该情况发生，需要低压断路保护器及时切断电路，保护硬件以及人员的安全。

## 1. 电流过大的原因

- 过负荷：并联过多用电器会导致干路电流过大。
- 短路：没有用电器消耗电能，导致电能全部转化为热能引起使电路烧毁。

## 2. 低压断路器

低压断路器在线路中起到过负荷和短路保护。长延时过电流脱扣器是热脱扣器，它具有反时限特性，也就是说过负荷的电流越大，其自动切断电源的时间也越短，反之，结果相反；瞬时过电流脱扣器是电磁脱扣器，当电流大于瞬时过电流脱扣器整定值，可以在约定的时间内切断电源。

## 3. 低压断路器的工作条件

**约定：出现预期的过电流故障时，断路器能够在约定时间内自动切断电源。**

- 说明：如果断路器的灵敏度差，当发生故障时，断路器不能在约定的时间内自动切断电源，若线路和设备在较长时间内通过较大的电流，会对其绝缘造成损害，降低设备和线路的使用寿命，同时对人身和设备安全产生威胁，严重的情况下可能会发生火灾等事故。

## 4. 计算方法

**《低压配电设计规范》GB50054-2011中规定：当短路保护电器为断路器时，被保护线路末端的短路电流不应小于断路器瞬时或短延时过电流脱扣器整定电流的1. 3倍。**

比如对于C32的断路器，其长延时过电流脱扣器的整定值或者额定值In为32A，脱扣曲线为C曲线，其瞬时脱扣范围为5In~10In。当故障电流大于10In，即320A时，断路器才会在约定时间内切断电源，考虑到断路器的制造误差和计算误差，要求保护线路末端的短路电流要不小于断路器瞬时或短延时过电流脱扣器整定电流的1.3倍。所以，当线路的断路器选择C32时，其保护线路末端的短路电流要不小于1.3×32×10=416A。对于线路来说，越靠近负荷端，其短路电流越小。

短路类型可以分为三相短路，两相短路，两相接地短路和单相接地故障，其中，单相接地故障短路电流最小。**如果当线路末端发生单相接地故障时，这种情况下短路电流最小，断路器都可以灵敏的动作，那么在发生其它情况下短路，断路器均可在约定时间内切断电源。**

---

举例说明：

在TN-S系统中，断路器的型号为C32，线路型号为全塑铜芯电缆5×10mm2，求在保证断路器灵敏度前提下，可以保护线路的最大长度。

1. 单相接地时的电路等效于220V交流电通过相线和地线后进入大地。对于较小截面的电缆的电阻远大于电抗，故可以把电抗忽略不计。全塑铜芯电缆10mm2的电阻每km为2.631Ω。

2. 选择用C32断路器，可以使该断路器在约定时间内可靠动作的最小短路电流为1.3×32×10=416A。

3. 线路发生单相接地故障时，设线路长度为L，则相保回路的电阻为2×2.631L=5.262LΩ。

4. 保证断路器的灵敏性需要满足：220/（5.262L）≧416，整理得到线路最大长度L为100m。
