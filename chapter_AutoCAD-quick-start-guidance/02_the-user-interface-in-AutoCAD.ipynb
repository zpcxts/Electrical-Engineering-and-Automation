# The user interface in AutoCAD

Navigate autoCAD like a pro with searchable commands, *hyper-customizable* workspaces, and an expansive *toolkit*.

1. Type: tutorial
1. Length: 1 min.

AutoCAD’s user interface provides all the tools you need to work on a drawing.

By becoming familiar with the interface, you can *streamline* your workflow.

From here, you can create or open drawings and projects, save drawings, *export files*, *access* the program options, publish or print files, or close drawings and projects.

Next, notice the Quick Access Toolbar.

This customizable toolbar *contains* a set of frequently used commands. You can add and remove commands as needed, either from the *ribbon* or from the toolbar itself.

If you need assistance locating a command or tool, you can use the Search field.

To access your user profile, *expand* the Account drop-down.From here, you can review account details, manage licenses, and sign out.

Expand the Help menu to access AutoCAD’s help content, download offline help or language packs, *adjust* privacy settings, and more.

The AutoCAD ribbon provides a *compact* *palette* of all the tools necessary to create or *modify* your drawing.

Like tools are organized into *panels*, some of which expand to show more tools.

By *default*, it is *docked* *horizontally* at the top of the drawing area, and you can customize its *placement*.

With AutoCAD’s tabs, you have quick access to your currently open drawings. You can also easily create new drawings, open preexisting drawings, and save or close all current drawings. 

Open the Start tab to review a *variety* of *initial* actions, including opening or creating drawings, access to recent Autodesk Projects, learning options, and announcements.

Now, open a blank drawing. Notice the *viewport* controls. Primarily used for 3D modeling, these controls provide a *convenient* way of changing views, visual styles, and other settings.

Next, notice the ViewCube. Displayed when you are working in 2D model space or 3D visual style, the ViewCube is a navigation tool that lets you adjust the *orientation* of viewing your drawing. The ViewCube can be moved to different locations within the viewport.

Now, review the navigation bar. Here, you can access navigation tools, such as the ViewCube and SteeringWheels. The user *coordinate* system (UCS) *establishes* the location and orientation of a movable *Cartesian* coordinate system.

By default, the UCS icon *appears* in the lower-left *corner* of the drawing area for the current model viewport. The dockable, *resizable* Command window accepts commands and system *variables*. It also displays *prompts* that help you complete a command sequence. This includes commands that were initiated at another location such as the ribbon.

In AutoCAD, a layout is a 2D working environment for creating drawing sheets. You can access one or more layouts from the layout tabs. Multiple layout tabs can be used to display details of the *various* *components* of your model at several *scales* and on different sheet sizes.

The *Status bar* displays the *cursor* location, drawing tools, and tools that affect your drawing environment. It provides immediate access to some of the most commonly used drawing tools.

You can also *toggle* settings such as *grid*, *snap*, *polar tracking*, and *object snap*.

AutoCAD is a robust application that puts the tools you need just a click away. And, with customization, you can create the workspace of your dreams.
