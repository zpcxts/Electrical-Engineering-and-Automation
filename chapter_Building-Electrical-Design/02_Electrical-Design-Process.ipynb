# 电气设计流程

## 一. 设计流程

- 设计委托书
- 设计院开始设计
- 审图机构审查图纸
- 把审查合格的施工图交付给甲方
- 施工单位开始施工

### 1. 设计委托书

一般是甲方委托设计人员，以甲方的口吻写出设计委托书，然后交由甲方盖章。设计委托书内容应该包括：

- 供电方案：变配电室数量，位置，变压器容量，辐射方式，各种（说的越细越好）。
- 针对单体的方案：哪个楼里放哪些东西，巴拉巴拉。
- 和其他专业配合的部分：针对不同专业提的条件，配合，修改。

一般，甲方会有自己的设计委托书模板。

### 2. 设计院开始设计

各个专业配合，然后确定本专业的方案。

### 3. 审图机构审查图纸

**审图机构审查设计图是否违反规范中的强条件**

一定会有修改建议，一般源自对规范理解不同的部分。有时会告诉你违反了规范中的强条件。反正就是和审图机构的审图老师反复沟通。

### 4. 把审查合格的施工图交付给甲方

甲方做标的（预算），做招投标，确定施工单位

### 5. 施工单位开始施工

施工单位基于图纸，和设计人员沟通施工困难的部分，图纸错漏空缺的部分。反正就是设计人员和施工人员反复沟通。
