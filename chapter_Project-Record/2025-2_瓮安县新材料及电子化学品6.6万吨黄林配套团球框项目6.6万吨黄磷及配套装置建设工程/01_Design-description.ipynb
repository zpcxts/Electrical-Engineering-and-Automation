# 设计说明

本笔记将会解释设计说明中的各个名词。并且说明出现这些名词，对于照明设计来说意味着什么。

设计说明应包含以下几个部分：
1. 工程概况
2. 设计依据
3. 涉及范围
4. 设计中涉及的各部分系统

## 一. 设计依据

### 1. 建筑概况
项目名称,建筑面积，层数，建筑高度，室内外高差，*建筑耐火等级*，*建筑结构形式*，*火灾危险性*，（后接详见建筑特征表）。*抗震设防烈度*，*室外消防用水量*。

1. **建筑耐火等级**：建筑的耐火等级是根据材料的耐火极限来划分的，一般分为 一级（最高）~四级（最低）。

乙级耐火等级（通常对应二级耐火等级）意味着：
1. 主要承重结构是钢筋混凝土或钢结构，耐火极限较高。
1. 屋面和楼板耐火极限 ≥ 1.0小时。
1. 墙体耐火极限 ≥ 2.5小时。
1. 一般用于中型工业厂房、商场等场所。

耐火等级影响照明设计，因为一些灯具可能要求防火外壳或特殊布线方式。

---

2. **钢筋混凝土框架结构**：框架结构指的是由柱、梁、楼板组成的承重体系，墙体主要是维护作用，不参与承重。

钢筋混凝土框架意味着：
1. 抗震能力较强，适合多层工业厂房。
1. 可以设计大跨度，适用于厂房、大型仓库。
1. 梁柱节点是关键部位，设计照明时要注意避开这些结构，以免影响布线和灯具安装。

---

3. **火灾危险性为甲类**：甲、乙、丙、丁、戊类是《建筑设计防火规范》对不同生产场所的火灾危险性分类。

甲类火灾危险性最高，意味着该厂房可能：
1. 处理或储存易燃易爆气体、液体（如氢气、乙炔、甲烷）。
1. 存在高温作业（如电炉炼钢）。
1. 需要特殊防火设计，如：
   - 防爆灯具（防止火花）。
   - 防爆开关（防止电弧引燃）。
   - 线路防护（如穿金属管）。

---

4. **抗震设防烈度6度**：意思是需要保证该建筑遇到6度（中等强度地震）也**不会塌陷**（但可能有一定的损坏）。

抗震设防烈度 6 度，意味着：
1. 该地区的地震强度可能达到 6 度（中等强度地震）。
2. 建筑需要按照 6 度设防，以确保结构在 6 度地震下不会倒塌，但可能有一定的损坏。
3. 照明系统设计要考虑抗震：
   - 灯具固定方式：避免坠落。
   - 线路敷设方式：减少地震时的损坏风险。
   - 应急照明：确保地震后仍有足够照明，便于人员撤离。

---

5. **室外消防用水量 35 L/S**：意为35升/秒

室外消防用水量指的是：
1. 火灾发生时，外部消防系统需要提供的最低水流量。
1. 35 L/S 说明该厂房需要的消防水量较大，属于火灾危险性较高的厂房。
