# 一. 设计说明

## 1. 消防应急照明和疏散指示系统

"非集中电源集中控制型系统"和"集中电源集中控制型系统"的区别与选择讲究：
1. 非集中电源集中控制型系统：
    - 电源分布：每个应急照明灯具或每个回路有独立的电源，通常是配备独立的应急电池。
    - 控制方式：控制部分是集中的，也就是控制器会集中管理各个灯具的开关、故障报警等。
    - 特点：适用于小范围或分散的应急照明场景，安装比较灵活，能够避免整个系统的单点故障。但是，由于电源分布，可能会增加维护和管理的复杂性。

2. 集中电源集中控制型系统：
    - 电源集中：所有应急照明灯具的电源由一个或几个集中的电源单元供给，通常配备大容量的蓄电池。
    - 控制方式：控制和电源部分也是集中管理，所有应急灯具由一个集中的控制系统进行监控和管理。
    - 特点：适用于较大规模的建筑或区域，安装和维护相对简便，因为电源和控制设备都集中在一个地方，便于统一管理。然而，一旦集中电源出现故障，可能影响整个系统的正常工作。

3. 选择时的讲究：
    - 场所大小：如果应急照明系统覆盖的区域比较大，使用集中电源集中控制型系统可能更合适，因为它能提供统一的电源和控制，简化管理。
    - 故障容忍度：如果对故障的容忍度较低，非集中电源的系统更能避免单点故障。
    - 成本与维护：集中电源系统可能需要更高的初期投资，但维护上相对简便；而非集中电源系统可能在电源设备上分散，但有更高的维护要求。
    - 建筑类型：一些特殊的建筑（如医院、数据中心等）可能会更倾向于使用非集中电源，以确保应急照明的独立性和可靠性。
