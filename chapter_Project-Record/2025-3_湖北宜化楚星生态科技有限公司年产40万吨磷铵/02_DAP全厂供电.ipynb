## 1.1 树脂浇注干式变压器

<p style="text-align:center;">
    <img src = "../../img/变压器01.png" alt="description" width="500"/>
</p>

- **树脂**：是一种难燃、阻燃、自熄的固体绝缘材料。
- **干式变压器**：指铁芯和绕组不浸渍在绝缘油中的变压器。
- **树脂浇注干式变压器**：环氧树脂干式变压器以环氧树脂为绝缘材料。高、低压绕组采用铜带（箔）绕成，在真空中浇注环氧树脂并固化，构成高强度玻璃钢体结构。绝缘等级有F、H级。环氧树脂干式变压器有电气性能好、耐雷电冲击能力强、抗短路能力强、体积小重量轻等特点。可安装温度显示控制器，对变压器绕组的运行温度进行显示和控制，保证变压器正常使用寿命。

## 1.2 电压互感器

变压器二次侧YY接法的小小变压器，即是电压互感器。电压互感器本质是变压器。

- 变压器：通过高低压变换，方便电力远距离传输。（通常容量大）
- 电压互感器：将高压变为低压，从而保护二次侧的仪表，控制电路。也保障检修人员的安全（通常容量小）
- 电压互感器不短路：电压互感器本身阻抗小，一旦副边短路，电流将激增而烧毁线圈。
- 保护方法：原边安装熔断器，副边可靠接地。以免原、副边绝缘层被击穿后，副边出现对地高电位而造成人身和设备事故。

- **这个设备是不是电压互感器存疑！！**，本处是业主要求安装的。

## 1.3 主断路器

- 括号表示这是一个可以抽插替换的原件
- 叉叉表示这是一个可以带负荷拉闸的开关

### 1.3.1 三段保护

三段式电流保护指的是电流速断保护（第一段）、限时电流速断保护（第二段）、定时限过电流保护（第三段），相互配合构成的一套保护。

- 第一段（瞬时）：当线路电流超过额定电流10倍时，瞬间跳闸。
- 第二段（延时）：当线路电流超过额定电流6倍时，经过0.1s后跳闸。
- 第三段（定时）：当线路电流达到额定电流0.95倍时，经过15s后跳闸。

## 1.4 电流互感器

<p style="text-align:center;">
    <img src = "../../img/电流互感器示意图.png" alt="description" width="300"/>
</p>

电流互感器的一次侧串联在被测线路中，二次侧接电流表。电流互感器的作用是：
- 用小量程的电流表测量大电流。
- 将高压线路和测量回路隔离开，保证作业人员的安全。

本图中，三个并排的圈圈代表隔离高压线路的电流互感器，第二排的孤独圈圈代表测量电流用的电流互感器。

电流互感器二次侧不允许开路：
- 二次侧电流产生的反磁通势会和一次侧电流产生的正磁通势抵消。二次侧没有电流，反磁通势会消失。铁芯上的总励磁磁通势过大会烧毁铁芯。同时，二次侧的感应电动势会非常大（有可能击穿空气）
- 保护措施：铁芯和二次侧绕组应当接地。

## 1.5 浪涌保护器（Surge Protection Device）

浪涌保护器的作用是抑制危害产生后的不良后果：

1. 雷击造成的瞬态冲击过电压
2. 大功率设备投切时产生的瞬态冲击过电压
