# 系统图

系统图是平面图的补充。在平面图中只会展示大致走线、机柜位置等。而系统图会详尽标注各种在平面图上未标出的参数。本次防雷接地系统图由两个部分构成：
1. 照明配电箱系统图
2. 应急照明电源箱系统图

## 一. 照明配电箱系统图

### 1. 照明配电箱型号

首先，我们会在配电箱图框上方标注照明配电箱的型号,例如，本案中选用的【照明配电箱 M02-31A-AL-3101】：
- M02:可能表示配电箱系列或类型（例如，M系列表示某种类型的配电箱）。
- 31A:可能是指配电箱的 额定电流，例如31安培。
- AL:可能表示该配电箱的材料或外壳类型（如铝制外壳）。
- 3101：可能是该配电箱的具体型号或序列号。

### 2.1 关于需要系数的说明

需要系数应用在线路末端，需要系数是针对安装负荷的，而不是针对计算负荷。它标志的是用电设备组投入运行时，从电网中实际取用的功率必须考虑的一个综合系数。终端配电箱的出线回路合理选择需要系数值，才能计算出来的结果符合或者接近电气设备实际运行情况，避免上级电缆、断路器等取值偏大，造成浪费。需要系数的取值需要查表。对于一组负荷的台数越多，需要系数取值越小，台数越少，需要系数取值越大。一般同一组动力设备台数在3台及以下时，其需要系数取值为1。

### 3. 接入配电柜的导线

### 3.1 总导线
连接配电柜的总导线上标注3个参数：

1. *电源由配电室低压配电柜引来* ： 这个东西应该是根据现场实际情况，和其他各部协商之后决定的
2. *ZC-YJV-1-5X10 DN50* ：这是电缆的规格，其中：
   - ZC：这代表电缆的 阻燃 类型，即这是一种具有阻燃功能的电缆。通常用于需要防火安全的场所。
   - YJV：这代表电缆的 结构类型 和 绝缘材料。具体含义是：
      - Y：表示电缆的绝缘材料是 聚乙烯（PE），这是一种常见的电缆绝缘材料，耐高温、耐腐蚀。
      - J：表示电缆是 交联聚乙烯（XLPE）绝缘的，即电缆的绝缘层经过交联处理，增强了绝缘性能。
      - V：表示电缆的外护套是 聚氯乙烯（PVC），这种外护套材料具有良好的耐腐蚀性。
   - 1：表示电缆是 单芯（单根电线）。
   - 5X10：表示电缆的结构，具体是由 5根导线（即五芯）组成，每根导线的截面积是 10 mm²。常用于三相电力系统，其中三根是相线，剩下两根可能是零线和地线。
   - DN50：这是电缆的 管道直径，即该电缆是通过一个 50 mm 直径的管道进行敷设的。这可能是指电缆在管道中的外径，或者是它通过的管道的尺寸。
3. *CH3N-63C40/3P* ： 这是开关的 **型号** 和 **规格**，具体含义如下：
   - CH3N：这是开关型号的前缀，表示开关是 CH3N系列。这个系列通常是 塑壳断路器（MCCB，Molded Case Circuit Breaker）的型号。
   - 63：这是开关的 额定电流，表示该开关的额定电流是 63 A。也就是说，开关最大能够承受的电流是 63 安培。
   - C40：这部分描述了开关的 跳闸特性：
      - C型：表示开关的 脱扣特性，即开关的跳闸时间和电流之间的关系。C型适用于负载电流较为平稳的场合，比如照明、空调等设备，能承受一定的过载电流（5到10倍额定电流的短时间内过载不跳闸）。
      - 40：表示开关的 额定短路脱扣电流，即在出现短路的情况下，开关跳闸的电流值为 40 倍额定电流，即 40 × 63A = 2520A。这意味着如果电流超过2520安培时，开关会自动断开，避免设备损坏。
   - 3P：表示开关是 三极开关，即适用于 三相电 的电力系统。三极开关用于控制三相电的接通与断开。

### 3.2 关于选择导线的一些补充说明

对于选线（即电缆的选择），在没有特殊要求或约束的情况下，通常由设计人员根据以下几个基本原则来选择：

1. 电流承载能力
2. 电缆的类型和用途
3. 额定电流与电缆匹配
4. 环境因素
5. 安装方式

在没有特别需求的情况下，通常可以按照图纸或标准选择这些电缆和开关配置，因为它们已经满足了常规的电力要求。ZC-YJV-1-5X10 DN50 和 CH3N-63C40/3P 是比较常见且符合大部分应用的配置。

### 3.3 分支导线

内涵和总导线的标注是一样的，标注内容为导线的型号和开关的型号。**注意**：规范中指出，如果用电器会暴露在潮湿，室外，或其他不太友好的环境中，则导线应选用**单相带零线**线路；开关应选带**漏电保护功能**的开关。

本例中，分支导线选用了一下两种：
1. 环境友好的：CH3N-63C16/1P
2. 环境不甚友好的：CH3N-63C16/1P+N；30mA带漏电保护开关

### 5. 接出配电柜的导线

以本例做说明：
1. *-WL1* ： 一号线路
2. *L1.N.PE* ： 指线路的 相线 (L1)，零线 (N) 和 保护地线 (PE)。这三种线分别用于传输电力、接地保护和中性点连接。
3. *ZC-YJV-0.6/1KV-3×2.5* ：表示电缆的规格和类型，具体来说是一根三相阻燃型电缆，额定电压为0.6/1 kV，导线截面为2.5 mm²。
4. *DN20* ： 管道的规格，表示电缆穿管使用的管道直径为 20mm。通常用于保护电缆，避免外界的物理损伤。

对于*ZC-YJV-0.6/1KV-3×2.5*更具体的说明：
- ZC：这通常指的是电缆的种类。ZC 表示 阻燃型电缆。
- YJV：这是电缆的型号，代表 聚氯乙烯绝缘电力电缆（YJV 一般是交联聚乙烯绝缘的电缆）。
- 0.6/1KV：这是电缆的额定电压，表示该电缆适用于 0.6 kV（对地） 和 1 kV（相与相之间） 的电压。
- 3*2.5：表示电缆有 3根导线，每根导线的截面面积为 2.5mm²。这里的 3 表示三根电线，包括一根火线、零线和地线。
