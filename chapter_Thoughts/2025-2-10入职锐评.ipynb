# 2025-2-10 入职锐评

今天是无语Zpc。

现在是下午3点，距离我入职已经过去了7个小时，但我仍然无法展开任何工作甚至是布置一下我的工位。我已经百无聊赖的在工位上下起了象棋，手机早没电了，今天的下班打卡也没戏了，更惨的是没法刷地铁卡，这意味着我得从拓东路走回家，这大冷的天，想想这些头就隐隐作痛。

这一切一切，都不得不令我锐评一下**办公室**这个组织，虽然我看不上不懂技术的办公室文职人员也不是一天两天了，今天实在是再一次加深了我对他们的刻板印象。

其实整体上，入职的过程非常顺利，杨老师已经帮我安排好了工位，进组后的工作和学习任务；技术部很快送来了电脑和外设（插播一句，单位的外设真是我平生用过的最垃圾的外设，这一部分看来得自费上班了）；另外还有一些信息录入啊，巴拉巴拉的手续，也蛮顺利的。

然而，然而，however，しかし。

工位上没有插线板，这如何叫我布置工位？我是说，身为电气组的选手，我竟然真的没有一点办法把埋在墙里的电引出来，这好比一个厨子，他已经备好了一切食材调料、锅碗瓢盆，然后忙活了半天发现没有火。这是很令人无语的。

其实我们上午10点就发现了这个问题。之所以是"我们"，是因为当我发现"没有电"的问题后，立刻询问了同部门的小伙伴，以及杨老师。一圈询问下来得知现在是办公室在管理此事，于是告知了办公室后，办公室反馈说下午来领插线板。

现在是下午3点。

......

现在是下午4点，杨老师帮我弄来一块插线板。感谢杨老师。
