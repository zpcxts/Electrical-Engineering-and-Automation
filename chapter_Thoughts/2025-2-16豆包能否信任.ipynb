# 豆包能否信任

这篇文章的由头是李总中午吃饭的时候和我说起人工智能辅助办公的事情。恰好这一周以来部门的其他同事也和我讨论过大语言模型辅助办公的问题，诉求如下：

- 赵工：辅助中俄英文互译。（效果还行）
- 敖工：辅助查询规范时，查询结果不准确。
- 赵工（小）：辅助画图。（做不到）
- 胡工：响应过慢。（服务器垃圾）

作为研究生阶段大量接触人工智能的我，可以听出大家在有自己的大语言模型实操体会，但是外行终究说不到点子上，所以我想谈一谈大语言模型（豆包，deepseek）最不靠谱的地方：**广告**

---
## 1. 大语言模型扮演了什么角色？

2000年，百度作为Google的竞品横空出世，在中国影响巨大。至今，我们遇到问题时，仍习惯用"百度一下"来代指"上互联网查询解决方案"。我们可以把百度想象成一个索引目录非常先进的图书馆，只要我们在【百度一下】搜索框内输入我们的问题，就可以得到我们想要的答案。

没错，此情此景，是否好像今天我们遇到问题时，打开豆包，在"豆包一下"中输入我们的问题，然后人工智能马上能给我们提供一个解决方案？————这就是大语言模型今时今日扮演的最大角色，他替代了搜索引擎，用更加先进的技术降低了我们获取解决方案的成本。

可以说，今天的豆包，像极了昨天的百度。它利用先进的识别和生成技术，理解我们的问题后快速生成一个解决方案；此外，豆包还融合了搜索引擎的功能，除了豆包本身生成的方案之外，它还从互联网中寻找既存的解决方案。也就是说，我们可以把豆包看作一个强化版搜索引擎，它可以比百度更精准，更高效地寻找到问题的解决方案，或者获取知识。

---
## 2. 百度的失败

我把百度和豆包的功能与辉煌作类比，同样，我认为百度和豆包对用户而言存在相同的问题。如今，百度已然腐朽，人们获取信息再也不会真正"百度一下"了，比如：
- 找吃的：大众点评
- 找旅游攻略：小红书
- 找视频：b站，抖音
- 找解决方案：CSDN

有人说，百度的失败标志着搜索引擎的失败。我认为这是**恶之花结出恶之果**的报应，对比谷歌，我们可以轻易证明并不是搜索引擎这条路走不通了，而是百度的**广告竞价策略**恶行累累终于遭了报应。今天，百度一下再也搜不出想要的内容，只有铺天盖地的广告。用户也弃百度而去。

这，也就引出了我今天想要表达的观点：
- 大语言模型是否靠谱？
- 大语言模型会不会像百度一样变成广告满天飞的搜索工具？
- 从技术上如何看待这个问题？
- 我们应该怎么做？
